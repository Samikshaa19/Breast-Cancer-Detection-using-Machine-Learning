# Breast Cancer Detection using Machine Learning
## Overview
Breast cancer is one of the most common types of cancer in women. Early detection can significantly increase survival rates. This project utilizes Machine Learning (Naïve Bayes Classifier) to classify tumors as malignant or benign based on provided features.

## Machine Learning Concept
Machine Learning is a subset of AI that enables computers to learn from data. In this project, we use a **Naïve Bayes Classifier**, a probabilistic model based on Bayes' Theorem, to perform classification. This model is particularly effective for medical diagnosis applications.

## About the Dataset
The dataset used is the **Breast Cancer Wisconsin dataset** from `sklearn.datasets`, which contains:
- **30 Features** (such as mean radius, texture, perimeter, area, etc.)
- **2 Target Labels**: Malignant (cancerous) and Benign (non-cancerous)


```python
# Load the Breast Cancer dataset
data = load_breast_cancer()
label_names = data["target_names"]  # Class labels: Malignant & Benign
labels = data["target"]  # Target values (0 = Malignant, 1 = Benign)
feature_names = data["feature_names"]  # Feature names
features = data["data"]  # Feature data
```
